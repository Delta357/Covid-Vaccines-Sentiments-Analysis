# Scrapping News Articles

News Articles are scrapped using BeautifulSoup and New York Times API

## 1. Times of India

```python
## Importing necessary modules
from bs4 import BeautifulSoup
import requests
import pandas as pd

#Creating a Dataframe
df = pd.DataFrame()
```

```python
#Regex Library
import re
def remove_whitespace(text):  ## remove_whitespace(text) removes extra whitespace
    return re.sub(' +', ' ',text)
```

## Sputnik

```python
i=1
k=0
max_pages = 11
next_link2= requests.get('https://timesofindia.indiatimes.com/topic/sputnik-v-vaccine/news')

while i<=max_pages:
    
    soup_next = BeautifulSoup(next_link2.content)
    links = soup_next.select('div.content a')
    
    actual_links = [link['href'] for link in links]
    
    for link in actual_links:
        doc_link = 'https://timesofindia.indiatimes.com'+link
        
        get_doc_url = requests.get(doc_link)
        soup_doc = BeautifulSoup(get_doc_url.content)
        
        try:
            text = soup_doc.select('div.ga-headlines')
            text = text[0].get_text()
            text = text.replace('\n',' ')
            doc = remove_whitespace(text)
            df.loc[k,'DOC'] = doc
            
            date = soup_doc.select('div._3Mkg-')
            date = date[0].get_text()
            df.loc[k,'DATE'] = date[-24:]
            
        except:
            df.loc[k,'DOC'] = '???'
        
        
        k=k+1
    
    page_links = soup_next.select('div.pagination a') 
    #page_links2 = [link['href'] for link in page_links]
    try:
        next_link = 'https://timesofindia.indiatimes.com'+page_links[-1]['href']
        next_link2= requests.get(next_link)
        
        i=i+1
    except:
        i=i+1
        
    print('Page',(i-1),'Done')
```
