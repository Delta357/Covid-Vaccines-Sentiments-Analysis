# Cleaning Public Tweets 

```python
#Importing libraries

#Data Manipulation
import pandas as pd
#Natural Language Processing
import nltk
#Regex
import re
```

## Moderna Tweets

```python
moderna_january = pd.read_csv('Moderna_tweets_January.csv')
moderna_february = pd.read_csv('Moderna_tweets_February.csv')
moderna_march = pd.read_csv('Moderna_tweets_March.csv')
moderna_april = pd.read_csv('Moderna_tweets_April.csv')
moderna_may = pd.read_csv('Moderna_tweets_May.csv')
moderna_june = pd.read_csv('Moderna_tweets_June.csv')
moderna_july = pd.read_csv('Moderna_tweets_July.csv')
moderna_august = pd.read_csv('Moderna_tweets_August.csv')
moderna_september = pd.read_csv('Moderna_tweets_September.csv')
```

```python
#Head of the dataframe
moderna_january.head()
```

```python
# Retrieving the Text Column from the Tweets

moderna_january_tweets = moderna_january['Tweet']
moderna_february_tweets = moderna_february['Tweet']
moderna_march_tweets = moderna_march['Tweet']
moderna_april_tweets = moderna_april['Tweet']
moderna_may_tweets = moderna_may['Tweet']
moderna_june_tweets = moderna_june['Tweet']
moderna_july_tweets = moderna_july['Tweet']
moderna_august_tweets = moderna_august['Tweet']
moderna_september_tweets = moderna_september['Tweet']
```

```python
moderna_february_tweets[0]
```

```python
def clean_tweet(tweets):
    '''
    Helps in cleaning text by removing Tweets URL and @username , removing all characters except Alphabets/Numbers
    and Returns clean Tweets list
    '''
    tweet_clean=[]
    for i in range(len(tweets)):
        #Removing ',' with no space (done for 3,700)
        if re.search(",", tweets[i]):
            result = re.sub('[,]','',tweets[i])
        else:
            result = tweets[i]
        #Replacing https with empty string
        result = re.sub(r"http\S+","",result)
        #Replacing @username with empty string
        result = re.sub(r"@\S+","", result)
        #Replacing all with empty space except Alphabets and digits
        result = re.sub('[^a-zA-Z\d]',' ',result)

        tweet_clean.append(result)
    return tweet_clean
```

```python
#Getting clean tweets 

moderna_january_clean_tweets = clean_tweet(moderna_january_tweets)
moderna_february_clean_tweets = clean_tweet(moderna_february_tweets)
moderna_march_clean_tweets = clean_tweet(moderna_march_tweets)
moderna_april_clean_tweets = clean_tweet(moderna_april_tweets)
moderna_may_clean_tweets = clean_tweet(moderna_may_tweets)
moderna_june_clean_tweets = clean_tweet(moderna_june_tweets)
moderna_july_clean_tweets = clean_tweet(moderna_july_tweets)
moderna_august_clean_tweets = clean_tweet(moderna_august_tweets)
moderna_september_clean_tweets = clean_tweet(moderna_september_tweets)
```

```python
# Covishield Clean Tweets List 
moderna_february_clean_tweets
```

```python
# Adding Tweet Clean Column in Dataframe 

moderna_january['Tweet_clean'] = moderna_january_clean_tweets
moderna_february['Tweet_clean'] = moderna_february_clean_tweets
moderna_march['Tweet_clean'] = moderna_march_clean_tweets
moderna_april['Tweet_clean'] = moderna_april_clean_tweets
moderna_may['Tweet_clean'] = moderna_may_clean_tweets
moderna_june['Tweet_clean'] = moderna_june_clean_tweets
moderna_july['Tweet_clean'] = moderna_july_clean_tweets
moderna_august['Tweet_clean'] = moderna_august_clean_tweets
moderna_september['Tweet_clean'] = moderna_september_clean_tweets
```
